# Visualização relação da estatisca de dados 
* a análise estatística é o processo de entender com variáveis no conjunto de dado que se relacionam uma com a outra e como essas relação dependem de outras variáveis. A Visualização de dados se torna uma peça essencial nesse processo, pois quando os dados são visualizados adequadamentes, as pessoas conseguem conseguir descobrir padrões que indicam relações nos dados.

# Função replot()
* metedo de alto nível para visualização de relação estatísticas entre variáveis. O métedo replot() ja combina com **scatter plot ** , e line plot 
* metedo usado para ve relação entre duas ou mais variaveis com uso dos parametros semanticos dos dados.

# Relação de variáveis com scatter plots 

### relação total da conta e da gorjeta 

```python
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', data=tips);
```

### Qualificando as cores
* separando as correlação por cores e diferenciamos o databreak;
* corelação do gráfico os pontos laranjas são de pessoas fumantes que efeturaram a gorjeta
* ponto azul totalmente qualificado para aqueles que não são fumantes e não fizeram a gorjeta 

```python
sns.relplot(x='total_bill', y='tip', hue='smoker' ,data=tips);
```

# comparando com os sexos ambos M e S 

```python
sns.relplot(x='total_bill', y='tip', hue='sex' ,data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex' , style='sex'  , data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex' , style='smoker'  , data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='size' , data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', size='size' , data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', size='size', sizes=(15,300)  , data=tips);
```
