## Oque é bloxplot 
-  whiker plot na tradução seria 'gráficos de caixa' ou 'gráficos de bigodes'.
-  Útil para visualizar outilers e distruibuição dos dados.
-  Útil para ver a média dos dados.

# Visualiando distribuição de observação
- dataset muito grande sera dificil de visualizar categórias 
- É preciso sumarizar a distribuição dos dados e facilitar as visualização
- Podemos usar o gráfico do tipo Bloxplot para visualizar a distribuição das variáveis
-Esse tipo de gráfico é muito útil para Outillers

```python
import seaborn as sns
import matplotlib
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
tips = sns.load_dataset('tips')
```

## Bloxplot de dias da semana por total da conta 

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips);
```

## Gráficos de bloxplot dos dias por total da conta se a pessoa é fumante ou não.

```python
sns.catplot(x='day', y='total_bill', kind='box',  hue='smoker',  data=tips);
```

- Tipos de boxplot como foco maior na fomra dos dodos do que nos outillers
- esse tipo de gráfico é interresante quando temos grandes datases 

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips);
```

- o metedo stripplot permite plotar distribuição dos dados
- podemos combinar os dois graficos para ter informação


```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips);
sns.stripplot(x='day', y='total_bill', data=tips, color='gray');
```

### o mtedo catplot(), permite que usamos facetgrid, assim podemos combinar gráficos em uma única figura

```python
%matplotlib inline 
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='boxen', height=4);
```

# Violin Plot 
- gráfico que ideal para visualizar a distribuição de variavies 
- combinação de bloxplot com KDE
- funciona bem para distribuição com picos 
- permite uma visualização mais rica do que com boxplot normal 

# violin plot 

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips);
```

## Violin Plot horizontal

```python
sns.catplot(x='total_bill', y='day', kind='violin', data=tips);
```

## Usando o parametro hue para ter uma 3 variavel 

```python
sns.catplot(x='day', y='total_bill', hue='time', kind='violin', data=tips);
```

- é possivel dividir o violin plot com o parametro split 
- isso e util quando temos muitos dados para exibir 

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, kind='violin', data=tips);
```

- usando o parametro inner para preencher o violin plot com obs
- útil para destacar as partes que possuem mais dados 

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, inner='stick',palette='pastel', kind='violin', data=tips);
```

### jointplot (scatterPlot e histograma)
- quando queremos ver a distribuuição de duas variaveis podemos usar o jointplot 
- união de gráficos do tipo scatter plot e histograma, para duas variaveis 
- mostra o scatter plot para duas variaveis e um histograma para cada variavel separadamente.

```python
sns.jointplot(x='tip', y='total_bill', color='r', data=tips)
```

### Hexbin plots
- adequado para visualização de grandes conjuntos de dados
- tende a funcionar melhor com as cores em branco de fundo

```python
sns.jointplot(x='tip', y='total_bill', color='r', kind='hex', data=tips)
```

### Usando o kde para plotar mais de uma variavel

```python
sns.jointplot(x='tip', y='total_bill', color='r', kind='kde', data=tips) 
```
