# Cores no Seaborn 
-  a função que controla as cores no seaborn e ['color_palette'].
-  essa função fornece uma interface para gerar cores no seaborn. 
-  usamos esse metedo internamente ao usar o parametro palette.
-  o metedo color_palette(), recebe o nome da paleta de cores do seaborn ou um colormap do matplotlib .
- o metedo set_palette(), recebe os mesmo paramentro e é usado para definir o estilo padrão de todos os plots.
- as cores devem ser definidas levando em consideração, o tipo de dados que queremos mostrar. (quantitativo, sequencial, driving).

### paleta de cores nos dados quantitativos

```python
sns.palplot(sns.color_palette('Accent'))
```

### paleta de cores para dados divergentes 

```python
sns.palplot(sns.color_palette('PuOr'))
```

### paleta de cores para dados sequencias 

```python
sns.palplot(sns.color_palette('Blues'))
```

# Plotando dados categóricos 
-  Quando trabalhamos com variáveis categóricas termos que visualizar dados de forma diferente.
- O seaborn fornce o métedo catplot(), que já contem diversos tipos de gráficos embutidos. 
- Isso facilita, pois , você pode usar diferentes gráficos usando um mesmo métedo.

# Gŕaficos de barras ou também conhecidos como gráficos de colunas 

```python
sns.catplot(x='sex', kind='count', palette='Set2', data=tips);
```

# gráfico de barra horizontais usando a coluna day 

```python
sns.catplot(y='day', kind='count', palette='Set1', data=tips);
```

# Scatter plot com dados categórios 

```python
sns.catplot(x='day', y='total_bill', palette='Set2', data=tips);
```

# O parâmetro swarm evita sobreposição de pontos

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips);
```

# o parâmetro hue permite que adcionaremos uma terceira variável a nossa visualização

```python
sns.catplot(x='day', y='total_bill', kind='swarm', hue='sex', palette='Dark2', data=tips);
```

- O parametro order permite alteramos a ordem padrão das categorias que estão exibidas 
- Isso é útil quando temos mais um gráficos na mesma figura e queremos as mesmas ordens.

```python
sns.catplot(x='day', y='total_bill', kind='swarm', hue='sex', order=['Sat', 'Sun', 'Fri', 'Thur'], data=tips);
```

## Invertendo a visualização para plots horizontais 

```python
sns.catplot(x='total_bill', y='day', kind='swarm', data=tips);
```

# gŕaficos com regrssão 
- quanto temos muitas variáveis quantitativas em nossos dados e interresantes visualizar como estas se relacionam.
- Podemos visualizar essas informação com linhas de regressão.
- Com modelos de regressão simples é possivel checar se existe alguma correlação entre algumas variáveis.

```python
%matplotlib inline 
sns.lmplot(x='total_bill', y='tip', hue='sex', palette='Pastel1', data=tips);
```

#### usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica time 

```python
%matplotlib inline
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', palette='Set2', data=tips);
```

## Parâmetros scatter_kws e line_kws

```python
%matplotlib inline 
sns.lmplot(
    x='total_bill',
    y='tip',
    hue='smoker',
    palette='Set1',
    data=tips,
    scatter_kws={'s': 90, 'alpha': 0.5},
    line_kws={'lw': 4}
);
```

### segregando gráficos pelo o parâmetro col e pelo parâmetro row 

```python
%matplotlib inline
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', row='sex', palette='Set2', data=tips);
```
